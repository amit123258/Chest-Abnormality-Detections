```python
import cv2
import numpy as np

# Read the image
data = cv2.imread('normal.jpg')

# Resize the image to 150x150
resized_data = cv2.resize(data, (150, 150))

# Normalize the image to a range of [0, 1]
normalized_data = resized_data.reshape((1,150,150,3))


```

```python
from tensorflow.keras.models import load_model

# Load the model
model = load_model('model.h5')

# Summary of the model
model.summary()
```

```python
model.predict(normalized_data)
```
